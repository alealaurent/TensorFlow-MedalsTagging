```python
# Derivative Work: Stephen Houser & Mohammad Irfan [Bowdoin College, 2017]
# Copyright 2017 The TensorFlow Authors. All Rights Reserved.
#
# Licensed under the Apache License, Version 2.0 (the "License");
# you may not use this file except in compliance with the License.
# You may obtain a copy of the License at
#
#     http://www.apache.org/licenses/LICENSE-2.0
#
# Unless required by applicable law or agreed to in writing, software
# distributed under the License is distributed on an "AS IS" BASIS,
# WITHOUT WARRANTIES OR CONDITIONS OF ANY KIND, either express or implied.
# See the License for the specific language governing permissions and
# limitations under the License.
# ==============================================================================

from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import argparse
import sys
import os

import numpy as np
import tensorflow as tf

#This function loads the model that has been learned above
def load_graph(model_file):
    graph = tf.Graph()
    graph_def = tf.GraphDef()

    with open(model_file, "rb") as f:
        graph_def.ParseFromString(f.read())
    with graph.as_default():
        tf.import_graph_def(graph_def)

    return graph

#This function transforms a .jpg image file into its "tensor" form
def read_tensor_from_image_file(file_name, input_height=299, input_width=299,
				input_mean=0, input_std=255):
    input_name = "file_reader"
    output_name = "normalized"
    file_reader = tf.read_file(file_name, input_name)
    #Assuming .jpg file
    image_reader = tf.image.decode_jpeg(file_reader, channels = 3,
                                        name='jpeg_reader')
    float_caster = tf.cast(image_reader, tf.float32)
    dims_expander = tf.expand_dims(float_caster, 0);
    resized = tf.image.resize_bilinear(dims_expander, [input_height, input_width])
    normalized = tf.divide(tf.subtract(resized, [input_mean]), [input_std])
    sess = tf.Session()
    result = sess.run(normalized)

    return result

#This function reads the list of tags from the output_labels.txt file
def load_labels(label_file):
    label = []
    proto_as_ascii_lines = tf.gfile.GFile(label_file).readlines()
    for l in proto_as_ascii_lines:
        label.append(l.rstrip())
    return label

#This function tests each image from a test set of images
def test():
    model_file = "model/output_graph.pb"
    label_file = "model/output_labels.txt"
    input_name = "import/Mul"
    output_name = "import/final_result"

    input_height = 299
    input_width = 299
    input_mean = 0
    input_std = 255

    #Create an output file
    output_file = "test_results.csv"
    out_file_object = open(output_file, "wt")

    graph = load_graph(model_file)
    input_operation = graph.get_operation_by_name(input_name);
    output_operation = graph.get_operation_by_name(output_name);

    #Print column headers  
    labels = load_labels(label_file)
    #print("medal_file,", end='')
    out_file_object.write("file,")
    for l in labels:
        #print("{},".format(l), end='')
        out_file_object.write("{},".format(l))
    #print("ground_truth")
    out_file_object.write("ground_truth\n")

    for name in os.listdir("test_set"): #For each subfolder in test_set
        if name.find(".") >= 0: #Ignore hidden files like .DS_Store
            continue
        print("Folder name:", name)
        folder_name = os.path.join("test_set",name) #Create subfolder name
        for medal_file in os.listdir(folder_name): #For each file in that subfolder
            if medal_file.endswith(".jpg"):
                print("\t"+medal_file)
                t = read_tensor_from_image_file(os.path.join(folder_name, medal_file),
                                      input_height=input_height,
                                      input_width=input_width,
                                      input_mean=input_mean,
                                      input_std=input_std)

                with tf.Session(graph=graph) as sess:
                    results = sess.run(output_operation.outputs[0],
                              {input_operation.outputs[0]: t})
                results = np.squeeze(results) #results is an array of probabilities
                res = ",".join(map(str, results)) #map applies str to each probability in results
                #print("{},{},{}".format(medal_file, res, name))
                out_file_object.write("{},{},{}\n".format(medal_file, res, name))
    
    out_file_object.close()  

#Apply the model to the test_set
test() 
```

# Questions 1

Write Python codes that reads in the test_results.csv file and computes (a) the overall test accuracy (in %) over the whole test set and (b) the test accuracy for each type of image (e.g., child).    



```python
def test_accuracy(csv_file):
    #Read in file
    file_object = open(csv_file, "rt")
    all_text = file_object.read()
    file_object.close()
    lines = all_text.split("\n")
    #print(lines[1])
    
    #A. Compute overall test accuracy: 
        #Check whether category identified in ground_truth corresponds with highest probability 2 categories
            #Create dictionary assigning each probability value to its category
    
    accurate_count = 0
    headers = lines[0].split(",")
    categories = {}
    hit = {}
    miss ={}
    
    for i in range(1, (len(headers)-1)): #Assign hit and miss dictionary values to 0
        hit[headers[i]] = 0
        miss[headers[i]] = 0

    for line in lines[1:-1]: #Go line by line
        data = line.split(",")
        ground_truth = data[-1]
        #print(data[14])
        

        for i in range(1, (len(data)-1)): #Assign probability value to category header
            categories[headers[i]] = data[i]
        #print(categories)
        
        #Reverse sort dictionary keys
        sorted_categories = sorted(categories, key = categories.get, reverse = True)
        #print(sorted_categories)
        
        
        #Check whether highest value header matches ground_truth
        if sorted_categories[0] == ground_truth or sorted_categories[1] == ground_truth:
            
            #Log the number of accurate responses
            accurate_count += 1
            hit[ground_truth] += 1
        else:
            miss[ground_truth] += 1
    
        
    #Calculate overall percent accurate
   
    print("Overall test accuracy:", (accurate_count / (len(lines)-2) * 100), "%" )
    #print("Hit:", hit)
    #print("Miss:", miss)
    
    #B. Calculate accuracy for each category:
    print("Percent accuracy for each image type:")
    for header in hit:
        print(header,((hit[header] / (hit[header] + miss[header]))* 100), "%")
    
        
        
    

test_accuracy("test_results.csv")
```
